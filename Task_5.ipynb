# Exploratory Analysis

```python
sns.set_style('darkgrid')
plt.bar('Region','Sales',data=df)
plt.ylabel('Sales',size=20)
plt.xlabel('Region',size=20)
plt.title('Sales vs Region',size=25)
```

From the South Region the sales were done more followed by South,Central,west 

```python
plt.figure(figsize=(30,30))
sns.catplot('State',kind='count',data=df,palette='Set1',height=6,aspect=1.5)
plt.xticks(rotation=90)
plt.title('countplot of Statewise sales',size=15)
```

From the above countplot colifornia people were ordered products more in the store followed by New York,Texas.

```python
plt.figure(figsize=(15,15))
sns.catplot(x="State", y="Sales", kind="bar", data=df)
plt.xticks(rotation=90)
```

 From the above bar plot Wyoming people were did more sales followed by Vermont,Montana

```python
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n({:d} )".format(pct, absolute) 
```

```python
#Segmentwise analysis of profit
fig=plt.figure(figsize=(10,7))
explode = (0.1, 0.1, 0.1,0.0)
colors = ( "orange", "cyan","grey","pink") 
data1= df.groupby(["Ship Mode"])['Sales'].aggregate(np.sum).reset_index().sort_values('Sales',ascending = False)
segment=data1["Ship Mode"].to_list()
data=data1["Sales"].to_list()

fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(data,  
                                  autopct = lambda pct: func(pct, data), 
                                  explode = explode,  
                                  labels = segment, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="black")) 
  
ax.legend(wedges, segment, 
          title ="Ship Mode", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("Ship Mode and Sales analysis") 
  
plt.show()
```

From the above plot the observation is that from the copiers.The store get more profit compared all the products and the tables are not purchased from the store.

```python
#Segmentwise analysis of profit
fig=plt.figure(figsize=(10,7))
explode = (0.1, 0.1, 0.1)
data1= df.groupby(["Segment"])['Profit','Sales'].aggregate(np.sum).reset_index().sort_values('Profit',ascending = False)
segment=data1["Segment"].to_list()
data=data1["Profit"].to_list()

fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(data,  
                                  autopct = lambda pct: func(pct, data), 
                                  explode = explode,  
                                  labels = segment, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="black")) 
  

ax.legend(wedges, segment, 
          title ="Segments", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("Segmentwise profit analysis") 
  
plt.show()
```

```python
df.groupby('Segment')['Profit','Sales','Quantity','Cost'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
```

The Store get more profit on the products purshed by the Consumer segment compare with the others segements. 

```python
# Setting the figure size
plt.figure(figsize=(16, 5))
plt.plot(df['Cost'], df['Sales'], color='#b80045')
plt.xlabel('Cost')
plt.ylabel('Sales')
plt.xticks(rotation='vertical', size=8)
plt.title('Cost vs Sales',size = 18)
plt.show()
```

The above plot shows the relation between the sales depending on the cost.The store get more sales on the product cost less than 2500/-.

```python
fig=plt.figure(figsize=(15,15))
sale_Products = df.groupby(["Segment","Sub-Category"])['Profit'].aggregate(np.sum).reset_index().sort_values('Profit',ascending = False)
sns.barplot(x = 'Segment', y= 'Profit' ,hue='Sub-Category',data=sale_Products,palette='rainbow')
plt.title('profit  by Segment & Sub-Category wise')
plt.xlabel('Segment')
plt.xticks(rotation=0)
plt.ylabel('Profit')
```

```python
fig=plt.figure(figsize=(15,15))
sale_Products = df.groupby(["Category","Sub-Category"])['Profit'].aggregate(np.sum).reset_index().sort_values('Profit',ascending = False)
sns.barplot(x = 'Category', y= 'Profit' ,hue='Sub-Category',data=sale_Products,palette='rainbow')
plt.title('profit  by Category & Sub-Category wise')
plt.xlabel('Category')
plt.xticks(rotation=0)
plt.ylabel('Profit')
```
