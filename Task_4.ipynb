# Decision Tree. In NutShell.
classifier is just like a flowchart diagram with the terminal nodes representing classification outputs/decisions. Starting with a dataset, you can measure the entropy to find a way to split the set until all the data belonngs to the same class. 
There are several approaches to decision trees like ID3, C4.5, CART and many more. 
For splitting nominal valued datasets you can use the ID3 algorithm. 
You can use matplotlib library to visualize the tree data. Decision Trees are prone to overfitting, thus to avoid overfitting you can prune the decision tree by combining the adjacent nodes that have low information gain.

```python
import pandas as pd
import numpy as np
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing libraries in Python
import sklearn.datasets as datasets
import pandas as pd

# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data,iris.target, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)

print('Decision Tree Classifer Created')
```

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df,y,test_size=0.3)
y_pred = dtree.predict(X_test)
```

```python
y_test
```

```python
mse = np.square(np.subtract(y_test,y_pred)).mean()
mse
```

```python
rmse = np.sqrt(mse)
```

```python
rmse
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
print("accuracy_score is:-",accuracy_score(y_test, y_pred))
print("classification report:-","\n",classification_report(y_test,y_pred))
print("confusion matrix:-","\n",confusion_matrix(y_test,y_pred))
```

Above decision tree was long in length.To reduce it we have to do pruning the tree.if the tree is less lengthy we will understand it better.

```python
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="gini", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Visualize the graph
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

Let's check the accuracy of the decision tree(max_depth)

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",accuracy_score(y_test, y_pred))
print("classification report:-","\n",classification_report(y_test,y_pred))
print("confusion matrix:-","\n",confusion_matrix(y_test,y_pred))
```

The accuracy was less than the above but we will get the short decision tree than above decision tree.
